
## 9. Conclusiones Generales

El presente proyecto tuvo como objetivo desarrollar y evaluar modelos de machine learning para la predicción de default crediticio, utilizando tanto herramientas tradicionales (scikit-learn) como entornos distribuidos (PySpark). A partir del análisis realizado, se pueden establecer las siguientes conclusiones:

---

### 9.1 Desempeño del Modelo

El modelo Random Forest optimizado mediante validación cruzada mostró un desempeño sólido y consistente en el conjunto de prueba:

- Alta accuracy global (92%).
- Buen equilibrio entre precision y recall.
- F1-score superior a 0.80.
- Excelente capacidad discriminativa (ROC AUC ≈ 0.95).

Estos resultados evidencian que el modelo es capaz de separar adecuadamente clientes con alta probabilidad de incumplimiento de aquellos con bajo riesgo.

---

### 9.2 Importancia de Métricas Adecuadas

Se demostró que la accuracy, aunque útil como referencia general, no es suficiente en problemas de clasificación con posible desbalance de clases.

El uso conjunto de:

- Precision
- Recall
- F1-score
- ROC-AUC
- Curvas Precision-Recall

permitió realizar una evaluación más robusta y alineada con las mejores prácticas en modelado predictivo.

Especialmente en riesgo crediticio, el recall adquiere gran relevancia, ya que minimizar los falsos negativos (defaults no detectados) es fundamental para reducir pérdidas financieras.

---

### 9.3 Impacto del Umbral de Decisión

El análisis del umbral de clasificación evidenció el trade-off natural entre precision y recall. Reducir el umbral incrementa la capacidad de detección de defaults, pero también aumenta el número de falsas alarmas.

Esto demuestra que el modelo no debe evaluarse únicamente con el umbral estándar (0.5), sino que el punto de operación debe definirse estratégicamente según el nivel de riesgo que la institución esté dispuesta a asumir.

---

### 9.4 Comparación entre Entornos (scikit-learn vs PySpark)

La implementación en scikit-learn presentó un mejor equilibrio de métricas en esta etapa experimental, probablemente debido a una optimización más detallada de hiperparámetros.

Por su parte, PySpark ofrece ventajas significativas en términos de escalabilidad y procesamiento distribuido, lo que lo convierte en una alternativa adecuada para entornos productivos con grandes volúmenes de datos.

Ambos enfoques demostraron ser válidos, pero su elección dependerá del contexto operativo.

---

### 9.5 Interpretabilidad del Modelo

El uso de LIME permitió aproximar el comportamiento del modelo en predicciones individuales, identificando variables relevantes como:

- Tasa de interés.
- Relación deuda-ingreso.
- Ingreso anual.
- Historial crediticio.

Esto aporta transparencia al sistema y mejora la confianza en la toma de decisiones basada en modelos automatizados.

---

### 9.6 Conclusión Final

El modelo desarrollado demuestra que técnicas de aprendizaje supervisado, adecuadamente optimizadas y evaluadas, pueden ofrecer herramientas robustas para la gestión del riesgo crediticio.

El análisis integral realizado confirma que:

- El modelo es estadísticamente sólido.
- Presenta alta capacidad de discriminación.
- Puede adaptarse estratégicamente mediante ajuste de umbral.
- Es interpretable en predicciones individuales.

En consecuencia, el sistema propuesto constituye una base confiable para aplicaciones de scoring crediticio y apoyo a la toma de decisiones financieras.
