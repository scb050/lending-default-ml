# 8. Comparación y Evaluación de Modelos

---

## 8.1 Contexto de Evaluación

En este capítulo se realiza la evaluación cuantitativa de los modelos entrenados para la predicción de default crediticio. 

Dado que el problema corresponde a una *clasificación binaria con posible desbalance de clases*, no es suficiente evaluar únicamente la métrica de accuracy. En escenarios de riesgo crediticio, es fundamental analizar la capacidad del modelo para detectar correctamente los casos de incumplimiento (default), ya que estos representan el mayor impacto financiero.

Por esta razón, se utilizan múltiples métricas que permiten una evaluación integral del desempeño del modelo.

---

## 8.2 Métricas de Evaluación Utilizadas

Para evaluar el rendimiento del modelo se emplearon las siguientes métricas:

- *Accuracy:* Proporción total de predicciones correctas sobre el total de observaciones.
- *Precision:* Proporción de predicciones positivas correctas sobre todas las predicciones positivas realizadas.
- *Recall (Sensibilidad):* Proporción de verdaderos positivos detectados sobre el total de positivos reales.
- *F1-score:* Media armónica entre precision y recall, utilizada cuando se busca un equilibrio entre ambas.
- *ROC AUC:* Área bajo la curva ROC, que mide la capacidad del modelo para discriminar entre clases a través de todos los posibles umbrales.

El uso de múltiples métricas permite evitar conclusiones erróneas derivadas del uso exclusivo de accuracy, especialmente en contextos donde las clases no están perfectamente balanceadas.

---

## 8.3 Resultados del Modelo Random Forest (scikit-learn)

El modelo entrenado y optimizado mediante validación cruzada con GridSearch presentó el siguiente desempeño en el conjunto de prueba:

- *Accuracy:* 0.9204  
- *Precision:* 0.7818  
- *Recall:* 0.8337  
- *F1-score:* 0.8070  
- *ROC AUC:* 0.9511  

### Interpretación de los Resultados

El modelo alcanza una accuracy superior al 92%, lo que indica un alto nivel de clasificación global correcta.

Sin embargo, las métricas más relevantes en este contexto son:

- El *Recall (83.37%)*, que muestra una alta capacidad para detectar clientes en default.
- La *Precision (78.18%)*, que indica que la mayoría de las alertas de default generadas por el modelo son correctas.
- El *F1-score (0.8070)*, que refleja un buen equilibrio entre detección y confiabilidad.
- El *ROC AUC (0.9511)*, que evidencia una excelente capacidad discriminativa entre clientes cumplidos e incumplidos.

En términos prácticos, el modelo logra identificar adecuadamente la mayoría de los casos de riesgo sin generar un número excesivo de falsas alarmas.

---

## 8.4 Análisis del Umbral de Decisión

El modelo clasifica por defecto utilizando un umbral de 0.5 sobre la probabilidad estimada de default.

No obstante, en problemas de clasificación desbalanceada, modificar este umbral puede alterar significativamente el equilibrio entre precision y recall:

- Reducir el umbral aumenta el recall (detecta más defaults).
- Sin embargo, disminuir el umbral también reduce la precision (más falsos positivos).

Este comportamiento refleja el trade-off natural entre sensibilidad y especificidad. 

En aplicaciones de riesgo crediticio, suele ser preferible priorizar un mayor recall, ya que dejar pasar un default real (falso negativo) puede tener mayores consecuencias financieras que generar una alerta adicional (falso positivo).

## 8.6 Comparación General entre scikit-learn y PySpark

A partir de los resultados obtenidos, se observa que:

- La implementación en *scikit-learn* presentó un mejor equilibrio entre precision y recall.
- El modelo en *PySpark* mostró mayor sensibilidad al umbral de decisión.
- Ambos enfoques logran separar adecuadamente las clases, pero el modelo optimizado con GridSearch en sklearn presentó métricas más robustas en el conjunto de prueba.

Desde una perspectiva técnica:

- scikit-learn resulta más flexible para experimentación y ajuste fino.
- PySpark es más adecuado para escenarios de Big Data y producción distribuida.

## 8.7 Conclusiones del Capítulo

El modelo Random Forest optimizado mediante validación cruzada demostró:

- Alta capacidad discriminativa (ROC AUC > 0.95).
- Buen equilibrio entre detección de default y control de falsas alarmas.
- Robustez frente a posibles desbalances de clase.
- Flexibilidad para ajustar el umbral según el nivel de riesgo deseado.

La evaluación integral mediante múltiples métricas permite concluir que el modelo desarrollado es adecuado para aplicaciones de scoring crediticio y puede servir como base para un sistema de apoyo a la toma de decisiones financieras.
