## 1.1 Contexto

El riesgo crediticio representa uno de los principales desafíos en el sector financiero, ya que la correcta identificación de clientes con alta probabilidad de incumplimiento (default) es fundamental para reducir pérdidas económicas y mejorar la estabilidad de las instituciones.

En la actualidad, las técnicas de Machine Learning permiten desarrollar modelos predictivos capaces de analizar grandes volúmenes de datos históricos y detectar patrones asociados al comportamiento de pago de los clientes.

En este proyecto se desarrolla un sistema de predicción de default crediticio utilizando técnicas de aprendizaje supervisado, con el fin de evaluar la capacidad de distintos enfoques computacionales para identificar clientes con alto riesgo financiero.

Se implementaron modelos utilizando tanto herramientas tradicionales de análisis (scikit-learn) como entornos distribuidos (PySpark), permitiendo comparar desempeño, escalabilidad e interpretabilidad.

## 1.3 Objetivo del Trabajo

### Objetivo General

Desarrollar y evaluar modelos de machine learning para la predicción de incumplimiento crediticio (default), comparando diferentes entornos de implementación y analizando su desempeño mediante métricas robustas de clasificación.

### Objetivos Específicos

- Implementar un modelo de clasificación supervisada para la detección de default.
- Optimizar hiperparámetros mediante validación cruzada.
- Evaluar el desempeño del modelo usando métricas como Precision, Recall, F1-score y ROC-AUC.
- Analizar el impacto del umbral de decisión en el equilibrio entre detección de riesgo y falsas alarmas.
- Aplicar técnicas de interpretabilidad para comprender la influencia de las variables en las predicciones.
- Comparar el desempeño entre scikit-learn y PySpark.
